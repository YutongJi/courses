```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt

for n in (1,5,10,20,30,50):
    plt.figure(figsize=(4,3))
    x = np.linspace(0, 1, 50)
    y=x**n/(x+5)
    plt.plot(y)
    print('when n is',n)
    plt.show()
```

```python
from scipy import  integrate  
x2 = lambda x: x**10/(x+5)
x3=integrate.quad(x2, 0, 1)
y0=x3[0]#integrate to calculate for y0 using giving a and n0

a=int(input('Please enter a :'))
n0=int(input('Please enter n0 :'))
n1=int(input('Please enter n1 :'))

if n1<n0:
    a=n1
    n1=n0
    n0=a
    
def fu(n1,n0,a):
    if n1 == n0:
        return y0
    else:
        #print(fu(n1-1,n0,a))
        return 1./int(n1)-a*fu(n1-1,n0,a)



fu(n1,n0,a)
```
