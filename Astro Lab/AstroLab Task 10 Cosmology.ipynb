# II. supernovae Ia as standard candle

```python
table3 = open('/Users/janet/Downloads/Cosmology_tab3.txt', 'r')
table4 = open('/Users/janet/Downloads/Cosmology_tab4.txt', 'r')
header1 = table3.readline()
header2 = table3.readline()
header3 = table3.readline()
header3 = table4.readline()
header1 = table4.readline()
header2 = table4.readline()

z=[]
z_error=[]
m_peak_obs=[]
delta_corr=[]
KBB=[]
AE=[]
for line in table3:
    line = line.strip()
    columns = line.split()
    z.append(columns[1])
    z_error.append(columns[2])
    m_peak_obs.append(columns[3])
    delta_corr.append(columns[7])
    KBB.append(columns[6])
    AE.append(columns[5])
z = np.asarray(z, dtype=np.float32)
m_peak_obs = np.asarray(m_peak_obs, dtype=np.float32)
delta_corr = np.asarray(delta_corr, dtype=np.float32)
KBB = np.asarray(KBB, dtype=np.float32)
AE = np.asarray(AE, dtype=np.float32)
z_error = np.asarray(z_error, dtype=np.float32)

m_b_eff=m_peak_obs + delta_corr -KBB-AE
logz=np.log10(z)
z_error=z_error*np.log(10)/z
#set_xlim(left=0.01, right=1.0)  
#print(z,logz)
plt.ylabel('effective mag m_eff')
plt.xlabel('redshift logz')
plt.plot(logz, m_b_eff,'b.')
plt.errorbar(logz, m_b_eff, yerr=z_error,linestyle='none')

z2=[]
z2_error=[]
m2_peak_obs=[]
delta2_corr=[]
KBB2=[]
AE2=[]
for line in table4:
    line = line.strip()
    columns = line.split()
    z2.append(columns[1])
    z2_error.append(columns[2])
    m2_peak_obs.append(columns[3])
    delta2_corr.append(columns[7])
    KBB2.append(columns[6])
    AE2.append(columns[5])
z2 = np.asarray(z2, dtype=np.float32)
m2_peak_obs = np.asarray(m2_peak_obs, dtype=np.float32)
delta2_corr = np.asarray(delta2_corr, dtype=np.float32)
KBB2 = np.asarray(KBB2, dtype=np.float32)
AE2 = np.asarray(AE2, dtype=np.float32)
z2_error = np.asarray(z2_error, dtype=np.float32)
print(z2_error)
z2_error=z2_error*np.log(10)/z2
m2_b_eff=m2_peak_obs + delta2_corr -KBB2-AE2
logz2=np.log10(z2) 
#print(logz2.size)
#print(m2_b_eff.size)
plt.ylabel('effective mag m_eff')
plt.xlabel('redshift logz')
plt.plot(logz2, m2_b_eff,'r.')
plt.errorbar(logz2, m2_b_eff,xerr=z2_error,linestyle='None')
x=logz
y=m_b_eff
def line(x, a, b):
    return a * x + b
popt, pcov = curve_fit(line, x, y)
yfit=line(x,popt[0],popt[1])
plt.plot(x,line(x,popt[0],popt[1]),'r')
plt.title('Hubble-Diagram with data from table3&4')
k=popt[0]
M=popt[1]
print('The slope is:',k)
print('M:', M)
#print(z2_error)
```
